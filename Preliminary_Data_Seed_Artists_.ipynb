```python
import pandas as pd
import numpy as np
import pickle
import time
```

```python
us_viral = pd.read_csv('viral-us-weekly-latest.csv')
us_weekly = pd.read_csv('regional-gb-weekly-latest.csv',header= 1)
gb_viral = pd.read_csv('viral-gb-weekly-latest.csv')
gb_weekly = pd.read_csv('regional-us-weekly-latest.csv', header= 1)
```

```python
#get artist column from each DataFrame
gb_viral_artists = list(gb_viral.Artist)
gb_weekly_artists = list(gb_weekly.Artist)
us_viral_artists = list(us_viral.Artist)
us_weekly_artists = list(us_weekly.Artist)
```

```python
#full artist list
full_artist_list = gb_viral_artists + gb_weekly_artists + us_viral_artists + us_weekly_artists
full_artist_list = list(set(full_artist_list))
```
